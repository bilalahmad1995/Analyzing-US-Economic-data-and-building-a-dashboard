```python
import pandas as pd
from bokeh.plotting import figure
from bokeh.io import show, output_notebook, push_notebook, output_file
from bokeh.layouts import row
```

```python
us_gdp = pd.read_csv('gdp.csv')
unemployment_data = pd.read_csv('clean_unemployment.csv')
```

```python
gdp_df = pd.DataFrame(us_gdp)
unemployment_df = pd.DataFrame(unemployment_data)
```

```python
#unemployment_greater_than_8.5percent
unemployment_greater = unemployment_df[unemployment_df['unemployment'] > 8.5]
```

```python
def make_dashboard(x, gdp_change, unemployment, title, file_name):
    output_file(file_name)
    output_notebook()
    p = figure(title=title, x_axis_label='year', y_axis_label='%')
    p.line(x.squeeze(), gdp_change.squeeze(), color="firebrick", line_width=4, legend="% GDP change")
    p.line(x.squeeze(), unemployment.squeeze(), line_width=4, legend="% unemployed")
    show(p)
```

```python
x = pd.DataFrame(gdp_df, columns=['date'])
```

```python
gdp_change = gdp_df['change-current']
```

```python
unemployment = unemployment_df['unemployment']
```

```python
title = "Unemployement stats according to GDP"

file_name = "index.html"
```

```python
make_dashboard(x, gdp_change, unemployment, title, file_name)
```
